# Abstract

Purpose: The purpose of this file, scraping_coffee_reviews, is to display the method used to scrape a coffee website (https://www.coffeereview.com) for information to be used for my Machine Learning Capstone project.  Each review contains information about one coffee bean, including identifiers such as:
 - name of the bean
 - location of the roaster
 - origin of the bean
 - date of the coffee review
 - price
 - the review
 - flavor/body/acidity/aftertaste/AGTRON/aroma scores
 
 
Findings: Scraping using BeautifulSoup was performed on April 3, 2020 and yielded 5515 coffee reviews starting in February, 1997 until April, 2020.

Signficance: The significance of the scraping is the ability to collect a non-standard dataset to perform supervised Machine Learning.

Conclusion: Collection of a unique dataset was successful.

# Scrape site for first of 3 list of URLS

```python
start = time.time()
coffees1 = []

# for index, url in enumerate(coffee_links):
for index, url in enumerate(coffee_links1):
        
    print("Loading page {} of {}".format(index + 1, len(coffee_links1)))
    coffee = {}
    r = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
    page = urlopen(r).read()
    soup = BeautifulSoup(page, 'html.parser')

    # Get each attritube using class and html tag and store in coffee dictionary
    coffee['rating'] = soup.find('span', attrs={'class':'review-template-rating'}).get_text()
    coffee['roaster'] = soup.find('p', attrs={'class':'review-roaster'}).get_text()
    coffee['title'] = soup.find('h1', attrs={'class':'review-title'}).get_text()

    # Find two tables of data
    tables = soup.find_all('table', attrs={'class':'review-template-table'})
    for table in tables:
        # Get entire table, and break it into rows
        table_rows = table.find_all('tr')
        for row in table_rows:
            # From each row, get each table data (first td is column name, second is data)
            tds = row.find_all('td')
            # store in dictionary column name = column data
            coffee[tds[0].get_text().rstrip(':')] = tds[1].get_text()

    # Get Bottom line part
    template = soup.find('div', attrs={'class':'review-template'})
    # Get all paragraphs in review template, then get last one
    bottom_line = template.find_all('p')[-1].get_text()
    # Store in dictionary
    coffee['bottom_line'] = bottom_line[bottom_line.index(':') + 1:].strip()

    # Get only text after bottom line :
#     try:
#         coffee['bottom_line'] = bottom_line[bottom_line.index(':') + 1:].strip()
#     except:
#         print("####################################")
#         print("####################################")
#         print(bottom_line)
        

    coffees1.append(coffee)

end = time.time()
print('Time taken: ', end - start)
```
