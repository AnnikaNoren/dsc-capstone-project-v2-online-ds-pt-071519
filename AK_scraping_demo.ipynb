```python
import requests
from bs4 import BeautifulSoup
```

```python
html_page = requests.get("https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/all-time")
```

```python
base_url = "https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/all-time/"
data = {}
for i in range(1,5):
    no = i*100 + 1
    new_url = base_url + str(no)
    html_pg = requests.get(new_url)
    data[new_url] = html_pg
```

```python
page = data["https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/all-time/101"]
```
