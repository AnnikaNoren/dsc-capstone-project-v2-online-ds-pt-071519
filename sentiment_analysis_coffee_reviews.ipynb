# Abstract

Purpose: The purpose of this page is to explore the relationship of the "Bottom Line" review given to each coffee bean and it's rating using Sentiment Analysis.

Findings: Each coffee review had a feature titled "Bottom Line" which was a short summary of the coffee bean (for example, "An unusual and evocative cup, paradoxically sweet and savory, with layered complexity.")  Sentiment analysis was performed on the text review using Twitter texts and restuarant reviews training models to see if a coffee review could be accurately classified as a postive sentiment or a negative sentiment.

Significance: The significance of sentiment analysis is to be able to predict a coffee's overall rating based on the "Bottom Line" review.

Conclusions:  Neither training model was able to accurately predict a positive/negative sentiment based on the "Bottom Line" review.  In the future, a model using actual coffee reviews would need to be used.

# Sentiment Analysis of Restuarant Reviews

```python
# https://www.geeksforgeeks.org/python-nlp-analysis-of-restaurant-reviews/
dataset = pd.read_csv('Restaurant_Reviews.tsv', delimiter = '\t')
```

```python
dataset.head(7)
```

```python
corpus = [] 
```

```python
for i in range(0, 1000):  
      
    # column : "Review", row ith 
    review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])  
      
    # convert all cases to lower cases 
    review = review.lower()  
      
    # split to array(default delimiter is " ") 
    review = review.split()  
      
    # creating PorterStemmer object to 
    # take main stem of each word 
    ps = PorterStemmer()  
      
    # loop for stemming each word 
    # in string array at ith row     
    review = [ps.stem(word) for word in review 
                if not word in set(stopwords.words('english'))]  
                  
    # rejoin all string array elements 
    # to create back into a string 
    review = ' '.join(review)   
      
    # append each string to create 
    # array of clean text  
    corpus.append(review)  
```

```python
# corpus
```

```python
# Creating the Bag of Words model 
from sklearn.feature_extraction.text import CountVectorizer 
  
# To extract max 1500 feature. 
# "max_features" is attribute to 
# experiment with to get better results 
cv = CountVectorizer(max_features = 1500)  
  
# X contains corpus (dependent variable) 
X = cv.fit_transform(corpus).toarray()  
  
# y contains answers if review 
# is positive or negative 
y = dataset.iloc[:, 1].values  
```

```python
# Splitting the dataset into 
# the Training set and Test set 
# from sklearn.cross_validation import train_test_split 
from sklearn.model_selection import train_test_split
  
# experiment with "test_size" 
# to get better results 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train
```

```python
# Fitting Random Forest Classification 
# to the Training set 
from sklearn.ensemble import RandomForestClassifier 
  
# n_estimators can be said as number of 
# trees, experiment with n_estimators 
# to get better results  
model = RandomForestClassifier(n_estimators = 501, 
                            criterion = 'entropy') 
                              
model.fit(X_train, y_train)  
```

```python
y_pred = model.predict(X_test) 
  
y_pred 
```
