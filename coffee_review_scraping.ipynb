## Scrape Coffee Reviews for all the names of coffee from their site

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
from pprint import pprint
import time
```

```python
coffee_links = []
```

```python
for i in range(1, 277):
    url = "https://www.coffeereview.com/review/page/{page_num}/".format(page_num=i)
    r = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
    page = urlopen(r).read()
    soup = BeautifulSoup(page, 'html.parser')

    # Find each paragraph with class called review-roaster (this is where link is)
    paragraphs = soup.find_all('p', attrs={'class':'review-roaster'})
    for paragraph in paragraphs:
        coffee_links.append(paragraph.find('a').get('href')) # 
        
# coffee_links
```
